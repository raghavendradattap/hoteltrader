```python
import pandas as pd
import numpy as np
import random
import string
from faker import Faker

fake = Faker()

# Number of records
n = 10000

# Variables
property_id = np.random.randint(1, 10001, n)
reservation_id = [''.join(random.choices(string.ascii_letters + string.digits, k=6)) for _ in range(n)]
booking_date = pd.date_range(start='2023-01-01', end='2023-08-15', freq='D').to_numpy()
booking_date = np.random.choice(booking_date, n)
check_in_date = pd.to_datetime(booking_date) + pd.to_timedelta(np.random.randint(0, 366, n), unit='D')
check_out_date_within_4 = check_in_date[:int(0.8*n)] + pd.to_timedelta(np.random.randint(1, 5, int(0.8*n)), unit='D')
check_out_date_4_to_8 = check_in_date[int(0.8*n):] + pd.to_timedelta(np.random.randint(4, 9, n-int(0.8*n)), unit='D')
check_out_date = np.concatenate([check_out_date_within_4, check_out_date_4_to_8])
guest_name = [''.join(random.choices(string.ascii_letters + string.digits, k=random.randint(13, 20))) for _ in range(n)]
number_of_guests = np.random.choice([2, 3], int(0.8*n))
number_of_guests = np.concatenate([number_of_guests, np.random.randint(1, 6, n-int(0.8*n))])
room_type = np.random.choice(['Suite', 'King', 'Queen', 'Double'], n)
payment_status = np.random.choice(['Paid', 'Unpaid'], n, p=[0.8, 0.2])
special_requests = np.random.choice(['Extra Bed', 'Late Check-Out', 'Early Check-In', 'Room with View', 'None'], n)
hotel_chain = np.random.choice(['Hilton', 'Wyndham', 'Radisson', 'Marriott', 'Hyatt', 'Accor', 'Independent'], n, p=[0.15, 0.15, 0.15, 0.15, 0.15, 0.15, 0.1])
hotel_name = [fake.company() for _ in range(n)]
hotel_location = [fake.city() for _ in range(n)]
marketplace = np.random.choice(['hotelbeds', 'webbeds', 'Expedia', 'Booking.com', 'Agoda', 'Sabre', 'Amadeus'], n)
rate_plan = np.random.choice(['Refundable', 'Non-Refundable', 'Inclusive of Breakfast', 'Half Board', 'Full Board'], n)
discount_code = [''.join(random.choices(string.ascii_letters + string.digits, k=8)) if random.random() < 0.2 else 'None' for _ in range(n)]
commission_rate = np.random.uniform(5, 20, n)
booking_source = np.random.choice(['Direct Website', 'OTA', 'Travel Agent', 'GDS', 'Wholesaler'], n)
booking_status = np.random.choice(['Confirmed', 'Cancelled'], n, p=[0.8, 0.2])
payment_method = np.random.choice(['Credit Card', 'Bank Transfer', 'Cash', 'Virtual Card'], n)
currency = np.random.choice(['USD', 'EUR', 'GBP', 'JPY', 'AUD', 'CAD'], n)
booking_channel = np.random.choice(['Direct', 'Partner', 'Development/Test', 'API', 'White Label'], n)
booking_reference = [''.join(random.choices(string.ascii_letters + string.digits, k=10)) for _ in range(n)]
room_rate = np.random.uniform(100, 450, int(0.8*n))
room_rate = np.concatenate([room_rate, np.random.uniform(200, 600, n-int(0.8*n))])
star_rating = np.random.choice([2, 3], int(0.5*n))
star_rating = np.concatenate([star_rating, np.random.choice([3.5, 4], int(0.4*n)), np.full(int(0.1*n), 5)])
room_rate = np.random.uniform(100, 450, int(0.8*n))
room_rate = np.concatenate([room_rate, np.random.uniform(200, 600, n-int(0.8*n))])
star_rating = np.random.choice([2, 3], int(0.5*n))
star_rating = np.concatenate([star_rating, np.random.choice([3.5, 4], int(0.4*n)), np.full(int(0.1*n), 5)])
room_rate[np.isin(star_rating, [4, 5])] *= 3
taxes_and_fees = np.random.uniform(5, 30, n)
contact_information = [fake.email() for _ in range(n)]
nationality = np.random.choice(['USA', 'European', 'Other'], n, p=[0.5, 0.3, 0.2])
loyalty_program = ['Yes' if random.random() < 0.2 else 'No' for _ in range(n)]
total_cost = room_rate * (pd.to_datetime(check_out_date) - pd.to_datetime(check_in_date)).days

# Recreating dataframe
data = pd.DataFrame({
    'Property ID': property_id,
    'Reservation ID': reservation_id,
    'Booking Date': booking_date,
    'Check-In Date': check_in_date,
    'Check-Out Date': check_out_date,
    'Guest Name': guest_name,
    'Number of Guests': number_of_guests,
    'Room Type': room_type,
    'Payment Status': payment_status,
    'Total Cost': total_cost,
    'Special Requests': special_requests,
    'Hotel Chain': hotel_chain,
    'Hotel Name': hotel_name,
    'Hotel Location': hotel_location,
    'Marketplace': marketplace,
    'Rate Plan': rate_plan,
    'Discount Code': discount_code,
    'Commission Rate': commission_rate,
    'Booking Source': booking_source,
    'Booking Status': booking_status,
    'Payment Method': payment_method,
    'Currency': currency,
    'Booking Channel': booking_channel,
    'Booking Reference': booking_reference,
    'Room Rate': room_rate,
    'Taxes and Fees': taxes_and_fees,
    'Contact Information': contact_information,
    'Nationality': nationality,
    'Loyalty Program': loyalty_program,
    'Star Rating': star_rating
})

data.head()

```
